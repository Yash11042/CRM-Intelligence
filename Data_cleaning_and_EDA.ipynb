```python
import pandas as pd
import numpy as np  
import seaborn as sns
import matplotlib.pyplot as plt
import  warnings
warnings.filterwarnings('ignore')

import os
import re
from collections import Counter
```

```python
#Loading the dataset
df = pd.read_excel(r"C:\Users\gopic\Downloads\1000 leads.xlsx")
df.head(3)
```

```python
#striping the column names
df.columns = [c.strip() for c in df.columns]
```

```python
#Normalize the 'Status' values

def normalize_status(s):
    if pd.isna(s):
        return None
    s2 = str(s).strip().lower()
    if 'convert' in s2:
        return 'Converted'
    if 'not' in s2 or 'no' in s2:
        return 'Not Converted'
    return s.strip().title()

df['status_cleaned'] = df['Status'].apply(normalize_status)

df.head()
```

```python
# Creating a assumptions Converted tp Interested and Not Converted to Not Interested

def interest_status(s):
    if s == 'Converted':
        return 'Interested'
    elif s == 'Not Converted':
        return 'Not Interested'
    else:
        return 'Unknown'
    
df['interest_status'] = df['status_cleaned'].apply(interest_status)

df.head()
```

```python
#Extracting the business executive names from 'Status Information

def extract_executive(s):
    if pd.isna(s):
        return 'Unknown'
    m = re.search(r'\(([A-Za-z0-9 _-]+)\)', str(s))
    if m:
        name = m.group(1).strip().lower()
        return name.title()
    return 'Unknown'

df['business_executive'] = df['Status information'].apply(extract_executive)

df.head()
```

```python
# Adding a new columns

keywords = ['interested' , 'demo' , 'later' , 'call' , 'details' , 'share' , 'info' , 'mail']

for kw in keywords:
    df[f'kw_{kw}'] = df['status information'].fillna('').str.lower().apply(lambda x : 1 if kw in x else 0)

```

```python
# Basic EDA

print("Dataset Shape:", df.shape)

print('\n')

print('columns:', df.columns.tolist())

print('\n')

print('Status_cleaned value counts:\n', df['status_cleaned'].value_counts())

print('\n')

print('Interest_status value counts:\n', df['interest_status'].value_counts())

print('\n')

print('Top 10 executives:\n', df['business_executive'].value_counts().head(10))
```

```python
#save the cleaned dataset

df.to_csv(r"C:\Users\gopic\Downloads\1000 leads cleaned.csv", index=False)
print("Cleaned dataset saved successfully.")
```
